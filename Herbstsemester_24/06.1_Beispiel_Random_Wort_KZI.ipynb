# Seed finden:

```python
import random

vokale = ["A","E","I","O","U"]
konsonanten = ["B","C","D","F","G","H","J","K","L","M","N","P","Q","R","S","T","V","W","X","Y","Z"]

x = 1

while True:
    random.seed(x)
    wort = random.choice(konsonanten)+random.choice(konsonanten)+random.choice(vokale)
    print(wort)
    if wort == "KZI":
        print(f"Der seed ist {x}!")
        break
    x += 1
```
