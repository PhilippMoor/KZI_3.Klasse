# Chunky Challenges: Shift Code, Mastermind, Passwords

Dieses Notebook enthaelt drei groessere Programmier-Challenges im Stil des Buchs. Schreiben Sie sauberen, gut kommentierten Code und testen Sie systematisch (inkl. Randfaelle).


### Challenge 1: Shift Code (Caesar)


Implementieren Sie eine **Shift-Codierung** (Caesar-Chiffre).

**Anforderungen**
1. Schreiben Sie eine Funktion `shift_encode(text, k)`.
   - `k` ist eine ganze Zahl (kann auch negativ sein).
   - Nur Buchstaben A–Z/a–z werden verschoben, alles andere bleibt unveraendert.
   - Gross/Kleinschreibung bleibt erhalten.
2. Schreiben Sie eine Funktion `shift_decode(text, k)`.
3. Schreiben Sie ein kleines Menue:
   - 1: Text codieren
   - 2: Text decodieren
   - 3: Ohne Schluessel: **alle 26 Varianten** ausgeben (Brute Force), damit man raten kann.

**Testen Sie** u.a.:
- k = 0, k = 26, k = -1, k = 53
- Texte mit Umlauten/Leerzeichen/Zahlen/Interpunktion


### Challenge 2: Mastermind (Farben)

Programmiere eine textbasierte Version von **Mastermind**.

**Spielregeln (vereinfachte Variante)**
- Der Computer wählt einen geheimen Code aus **4 Farben**, z.B. aus: `['Rot','Blau','Grün','Gelb','Orange','Violett']`.
- Eingabe eines Tipps: 4 Farbnamen (in gleicher Schreibweise).
- Rückmeldung pro Tipp: Anzahl Farben an richtiger Position (schwarz) und Anzahl richtiger Farben an falscher Position (weiss).
- Wiederhole, bis der Code gelöst ist oder Versuche aufgebraucht sind.

**Hinweis:** Du kannst die Farben als Liste speichern und per `random.choice` viermal auswählen.


### Challenge 3: Passwords


Bauen Sie ein kleines **Passwort-System** (Konsole). Ziel ist ein realistischer Umgang mit Regeln und Speicherung.

**Teil A: Passwort-Regeln**
Schreiben Sie `is_strong(password)` mit folgenden Regeln:
- Mindestens 10 Zeichen
- Mindestens 1 Grossbuchstabe, 1 Kleinbuchstabe, 1 Ziffer
- Mindestens 1 Sonderzeichen aus `!@#$%^&*()-_+=` (Sie duerfen die Menge erweitern)
- Darf **nicht** den Benutzernamen als Teilstring enthalten

**Teil B: Speichern (ohne Klartext)**
- Speichern Sie Passwoerter nicht im Klartext.
- Verwenden Sie `hashlib` mit Salt (z.B. `pbkdf2_hmac`).
- Speichern Sie pro Benutzer: Salt + Hash (z.B. in JSON-Datei).

**Teil C: Login und Sperre**
- Erlauben Sie Registrierung und Login.
- Nach 3 Fehlversuchen: 30 Sekunden Sperre (Timer).

**Zusatz (optional, schwierig):** Passwort-Reset mit Sicherheitsfrage oder einmaligem Reset-Code.
