### Aufgabe 1
Lies eine Temperatur in Celsius ein und rechne sie in Fahrenheit und Kelvin um.

Formeln:
- Fahrenheit = Celsius × 1.8 + 32
- Kelvin = Celsius + 273.15

Beispiel: 100 °C → 212.0 °F, 373.15 K

<details><summary>Lösung</summary>

```python
celsius = float(input('Temperatur in °C: '))
fahrenheit = celsius * 1.8 + 32
kelvin = celsius + 273.15
print(f'{celsius} °C = {fahrenheit} °F = {kelvin} K')
```

### Aufgabe 2

Gib alle Zahlen von 1 bis 50 aus, die gleichzeitig durch 3 **und** durch 5 teilbar sind.

Erwartete Ausgabe: 15, 30, 45

<details><summary>Lösung</summary>

```python
for i in range(1, 51):
    if i % 3 == 0 and i % 5 == 0:
        print(i)
```

### Aufgabe 3
Berechne mit einer for-Schleife die Summe aller geraden Zahlen von 2 bis 100 (inklusive).

Erwartetes Ergebnis: 2550

<details><summary>Lösung</summary>

```python
summe = 0
for i in range(2, 101, 2):
    summe += i
print(summe)
```

</details>

### Aufgabe 4

Lies in einer while-Schleife wiederholt Zahlen mittels input ein. Sobald die eingegebene Zahl negativ ist, brich ab und gib die Summe aller bisher eingegebenen (positiven) Zahlen aus.

Beispiel: 4, 7, 3, -1 → Summe: 14

<details><summary>Lösung</summary>

```python
summe = 0
while True:
    zahl = int(input('Zahl (negativ zum Beenden): '))
    if zahl < 0:
        break
    summe += zahl
print('Summe:', summe)
```

</details>

### Aufgabe 5
Gib eine vollständige Multiplikationstabelle von 1×1 bis 5×5 aus.

Format pro Zeile: `1 x 1 = 1`, `1 x 2 = 2`, ..., `5 x 5 = 25`

<details><summary>Lösung</summary>

```python
for i in range(1, 6):
    for j in range(1, 6):
        print(f'{i} x {j} = {i * j}')
```

### ★ Aufgabe 7
Lies eine Zahl n ein und berechne n! (n-Fakultät) mit einer for-Schleife.

Beispiel: 6 → 720  
Hinweis: n! = 1 × 2 × 3 × … × n

<details><summary>Lösung</summary>

```python
n = int(input('n: '))
fak = 1
for i in range(1, n + 1):
    fak *= i
print(f'{n}! = {fak}')
```

### Aufgabe 8
Würfle 100-mal mit einem 6-seitigen Würfel und zähle, wie oft jede Augenzahl (1–6) aufgetreten ist. Gib das Ergebnis übersichtlich aus.

Beispiel-Ausgabe:
```
1: 18x
2: 15x
...
```

<details><summary>Lösung</summary>

```python
import random

zaehler = [0] * 7  # Index 1-6 nutzen
for _ in range(100):
    wurf = random.randint(1, 6)
    zaehler[wurf] += 1

for auge in range(1, 7):
    print(f'{auge}: {zaehler[auge]}x')
```

</details>

### Aufgabe 9
Der Computer denkt sich eine zufällige Zahl zwischen 1 und 20. Der Spieler hat maximal 5 Versuche. Nach jedem falschen Tipp gibt das Programm aus, ob die gesuchte Zahl grösser oder kleiner ist.

<details><summary>Lösung</summary>

```python
import random

geheim = random.randint(1, 20)
gewonnen = False

for versuch in range(1, 6):
    tipp = int(input(f'Versuch {versuch}/5 – dein Tipp: '))
    if tipp == geheim:
        print(f'Richtig! Du hast es in {versuch} Versuch(en) geschafft.')
        gewonnen = True
        break
    elif tipp < geheim:
        print('Zu klein!')
    else:
        print('Zu gross!')

if not gewonnen:
    print(f'Leider verloren. Die Zahl war {geheim}.')
```

</details>

### Aufgabe 11
Gegeben ist das Wort `'Programmierung'`. Gib folgendes aus:
1. Die ersten 5 Zeichen
2. Die letzten 4 Zeichen
3. Jeden zweiten Buchstaben
4. Das Wort rückwärts

<details><summary>Lösung</summary>

```python
wort = 'Programmierung'
print(wort[:5])        # 'Progr'
print(wort[-4:])       # 'rung'
print(wort[::2])       # jeden 2. Buchstaben
print(wort[::-1])      # rückwärts
```

### Aufgabe 12
Schreibe eine Funktion `ist_bestanden(note, minimum)`, die `True` zurückgibt wenn `note >= minimum` und sonst `False`.

Nutze sie, um für eine Liste von Noten `[3.5, 4.0, 2.8, 5.1, 3.9]` auszugeben, ob jede Note bestanden (Minimum: 4.0) ist oder nicht.

<details><summary>Lösung</summary>

```python
def ist_bestanden(note, minimum):
    return note >= minimum

noten = [3.5, 4.0, 2.8, 5.1, 3.9]
for note in noten:
    status = 'bestanden' if ist_bestanden(note, 4.0) else 'nicht bestanden'
    print(f'Note {note}: {status}')
```

</details>

### Aufgabe 13
Schreibe eine Funktion `summe_bis(n)`, die die Summe aller ganzen Zahlen von 1 bis n berechnet und zurückgibt. Verwende dazu eine for-Schleife.

Beispiel: `summe_bis(10)` → 55

<details><summary>Lösung</summary>

```python
def summe_bis(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

print(summe_bis(10))   # 55
print(summe_bis(100))  # 5050
```

</details>
